## Data Pre-Processing 

### Data Pre-Processing English

```python
df_en.shape
```

The data has 300 rows and 85 columns.

```python
df_en.info()
```

```python
# Only get the important columns, since most columns values are null
df_en = df_en[['created_at','full_text','favorite_count','view_count','reply_count','retweet_count']] 
```

```python
df_en.describe() # Get statical description
```

```python
df_en.isna().sum() #Check for null values
```

```python
df_en.duplicated().sum() # Check for duplicates
```

```python
df_en['view_count'].fillna(value=df_en['view_count'].mean(), inplace=True) # Fill the missing values of view_count columns with mean
```

```python
df_en.isna().sum() #Check for null values
```

### Data Pre-Processing Arabic

```python
df_ar.shape
```

The arabic dataset have 300 rows and 61 columns

```python
df_ar.info()
```

```python
# Only get the important columns, since most columns values are null
df_ar = df_ar[['created_at','full_text','favorite_count','view_count','reply_count','retweet_count']] 
```

```python
df_ar.describe() # Get statical description
```

```python
df_ar.isna().sum() #Check for null values
```
